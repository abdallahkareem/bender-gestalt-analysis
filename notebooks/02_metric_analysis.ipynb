# 02 — Geometric Metric Analysis
Explore how Hu Moments, Hausdorff Distance, and Rotation Angle behave across our 89-image dataset.

```python
import sys
sys.path.append('..')

import json
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from src.preprocessing.pipeline import PreprocessingPipeline
from src.algorithm.comparator import ShapeComparator
```

```python
# Run comparison on all images
pipeline = PreprocessingPipeline()
comparator = ShapeComparator(template_dir='../data/templates/')

all_metrics = []
images = list(Path('../data/raw_drawings/').glob('*.png'))
print(f'Processing {len(images)} images...')

for img_path in images:
    preprocessed = pipeline.run(str(img_path))
    comparison = comparator.compare(preprocessed)
    all_metrics.append({'image': img_path.name, 'comparison': comparison})

print('Done.')
```

```python
# Plot Hu Similarity Distribution per Figure
figures = ['A','1','2','3','4','5','6','7','8']
hu_scores = {fig: [] for fig in figures}

for record in all_metrics:
    for fig in figures:
        score = record['comparison'].get(fig, {}).get('hu_similarity', None)
        if score is not None:
            hu_scores[fig].append(score)

fig_plot, axes = plt.subplots(1, 9, figsize=(20, 4), sharey=True)
fig_plot.suptitle('Hu Moment Similarity per BGT Figure (all 89 patients)', fontsize=13)

for ax, fig_label in zip(axes, figures):
    data = hu_scores[fig_label]
    ax.boxplot(data, patch_artist=True,
               boxprops=dict(facecolor='#4a90d9', alpha=0.7))
    ax.set_title(f'Fig {fig_label}', fontsize=10)
    ax.set_xticks([])
    ax.set_ylim(0, 1)

axes[0].set_ylabel('Hu Similarity Score')
plt.tight_layout()
plt.savefig('../assets/hu_similarity_boxplot.png', dpi=150, bbox_inches='tight')
plt.show()
```
