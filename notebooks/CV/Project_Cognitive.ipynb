```python
df = df.drop(columns=['رقم الحالة', 'الاسم'])
```

```python
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder , OneHotEncoder

le = LabelEncoder()
df['النوع'] = le.fit_transform(df['النوع'])
```

```python
df["استخدام الهامش"] = le.fit_transform(df["استخدام الهامش"])
df["موضع الرسم"] = le.fit_transform(df["موضع الرسم"])
df["تغيير موضع الورقة"] = le.fit_transform(df["تغيير موضع الورقة"])
df["المسافة بين رسمين"] = le.fit_transform(df["المسافة بين رسمين"])
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder_1 = OrdinalEncoder(categories=[["لا يوجد تداخل", "تداخل الرسوم", "تصادم أو نزعة إليه"]])
df[["الاتصال أو التصادم"]] = ordinal_encoder_1.fit_transform(df[["الاتصال أو التصادم"]])

ordinal_encoder_2 = OrdinalEncoder(categories=[["الرابع الإبتدائي", "السادس الإبتدائي", "الأول إعدادى", "الثالث إعدادى"]])
df[["الصف الدراسي"]] = ordinal_encoder_2.fit_transform(df[["الصف الدراسي"]])
```

```python
one_hot_cols = [
    "التصنيفات المرضية",
    "التغيرات في الجشطالت",
    "التسلسل",
    "ملاحظات دالة بشكل عام"
]

df = pd.get_dummies(df, columns=one_hot_cols)
```

Extract impportant info from text and drop it after

```python
from sklearn.feature_extraction.text import TfidfVectorizer

df["التعليق العام"] = df["التعليق العام"].fillna("")

tfidf = TfidfVectorizer(max_features=100)

tfidf_matrix = tfidf.fit_transform(df["التعليق العام"])

tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=tfidf.get_feature_names_out(),
    index=df.index
)

df = pd.concat([df, tfidf_df], axis=1)

df.drop("التعليق العام", axis=1, inplace=True)
```
